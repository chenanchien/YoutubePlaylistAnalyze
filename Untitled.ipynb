```python
import urllib.request
from bs4 import BeautifulSoup as bs
import locale
import re
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

```python
url = "https://www.youtube.com/playlist?list=PLwItru4bLdHx3nnUrFUBFWwMHuo_4Yx9P"
response = urllib.request.urlopen(url)
html_cont = response.read()
soup = bs(html_cont,"html.parser",from_encoding = "UTF-8")
print(soup.prettify())
```

```python
titlelist = []
urllist = []
timelist = []
infos = soup.find_all("td",class_="pl-video-title")
for info in infos:
    titlelist.append(info.a.text)
    urllist.append(re.findall(r"\?v=(.{11})",info.a["href"])[0])
times = soup.find_all("div",class_="timestamp")
for time in times:
    timelist.append(time.text)
```
